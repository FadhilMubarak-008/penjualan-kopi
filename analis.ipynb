untuk melihat naik turunnya penjualan

```python
df['date']=pd.to_datetime(df['date'])
penjaualan_harian = df.groupby('date')['money'].sum()
penjaualan_harian
```

```python
penjaualan_harian.plot()
```

cek kopi mana yg paling laris

```python
produk_penjualan=df.groupby('coffee_name')['money'].sum().sort_values(ascending=False)
produk_penjualan
```

```python
produk_penjualan.plot(kind='bar')
```

jumlah transaksi per produk

```python
jumlah_transaksi_product=df['coffee_name'].value_counts()
jumlah_transaksi_product
```

```python
jumlah_transaksi_product.plot(kind='bar')
```

Lihat harga setiap produk

```python
avg_price = df.groupby('coffee_name')['money'].mean()
avg_price
```
